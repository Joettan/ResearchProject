```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import pandas as pd

# Give min max std var histogram density diagram based on csv file
population_df = pd.read_csv("./data/population.csv",header=0)
def AvgData(df):
    for i in range (2001,2021):
        print("Year ",i," average value = ",df[str(i)].mean())
def Maximum(df):
    for i in range (2001,2021):
        print("Year ",i," Maximum value = ",df[str(i)].max())
def Minimum(df):
    for i in range (2001,2021):
        print("Year ",i," Minimum value = ",df[str(i)].min())
def Std(df):
    for i in range (2001,2021):
        print("Year ",i," Std value = ",df[str(i)].std())
AvgData(population_df)
Maximum(population_df)
Minimum(population_df)
Std(population_df)
```

```python
def AddAllData(df,start,end):
    result = df.iloc[:,start]
    for i in range(start+1,end):
        result = result.append(df.iloc[:,i],ignore_index=True)
    return result
result_df = AddAllData(population_df,2,23)
print(result_df)
```
